```python
import torch
from torch.utils.data import Dataset, DataLoader
```

```python
class DS(Dataset):
    def __init__(self):
        print('init')
        self.data = {'x': [[i , i + 1] for i in range(10)], 'xx': [i*2 for i in range(10)], 'xxx': [i*3 for i in range(10)], 'y': [i / 10 for i in range(10)]}

    def __len__(self):
        return len(self.data['x'])
    
    def __getitem__(self, index):
        return {'x': torch.tensor(self.data['x'][index]), 'y': torch.tensor(self.data['y'][index])}
```

```python
dl = DataLoader(DS(), batch_size=2, shuffle=True)
for e in range(10):
    itr = iter(dl)
    print(e)
    for i in itr:
        print(i)
```
